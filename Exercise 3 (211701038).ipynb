# Data Visualization

```python
plt.hist(df['Base_Salary'],color='green',edgecolor='red')
plt.title('Histogram',fontsize=20)
plt.show()
```

```python
df.boxplot('Bonus')
```

```python
plt.scatter(df['Base_Salary'],df['Net_Salary'])
plt.xlabel('Base Salary')
plt.ylabel('Net Salary')
plt.title("Scatter Plot")
plt.show()
```

```python
import numpy as np
import seaborn as sns
corr = df.corr()
mask = np.ones_like(corr)
sns.heatmap(corr,annot=True,cmap="vlag",vmin=-1,vmax=1)
plt.show()
```

# Advanced Data Visualization and Insights

```python
sns.pairplot(df,hue='Department')
plt.show()
```

```python
plt.bar(df['Department'],df['Net_Salary'].mean())
plt.show()
```

```python
grouped_data = df.groupby('Position')['Overtime_Hours'].sum().reset_index()
grouped_data
```

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['Position'], grouped_data['Overtime_Hours'])
plt.xlabel('Position')
plt.ylabel('Overtime Hours')
plt.show()
```
