```python
import pandas as pd
```

```python
dt = pd.read_csv('housing_dataset.csv')
dt
```

# Handling null values

```python
dt.isnull()
```

```python
dt['Price'] = dt['Price'].fillna(dt['Price'].median())
dt
```

```python
dt['Bathrooms'] = dt['Bathrooms'].fillna(dt['Bathrooms'].mean())
dt
```

```python
dt = dt.dropna()
dt
```

# Handling Outliers

```python
q1 = dt['Price'].quantile(0.25)
q3 = dt['Price'].quantile(0.75)
```

```python
IQR = q3 - q1
lb = q1 - (1.5*IQR)
ub = q3 + (1.5*IQR)
```

```python
lb
```

```python
ub
```

```python
outlier = ((dt[['Price']]<lb)|(dt[['Price']]>ub)).any(axis=1)
```

```python
cleaned_dt = dt[~outlier]
cleaned_dt
```

```python
#MinMax
min_sq = cleaned_dt['Sqft_Living'].min()
max_sq = cleaned_dt['Sqft_Living'].max()
```

```python
cleaned_dt['Sqft_Living'] = (cleaned_dt['Sqft_Living']-min_sq)/(max_sq-min_sq)
cleaned_dt
```

# Data Cleaning & Transformation

```python
cleaned_dt.dtypes
```

```python
cleaned_dt['Year_Built']=cleaned_dt['Year_Built'].astype(int)
cleaned_dt['Year_Built']
```

```python
dummies = pd.get_dummies(cleaned_dt,columns=['Waterfront'],drop_first='True')
dummies
```

```python
dummies.to_csv('New_housing_dataset.csv')
```
